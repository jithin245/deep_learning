```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
```

```python
# load an image
image_path = 'dog.jpeg'
original_image = load_img(image_path)
image_array = img_to_array(original_image)
image_array = np.expand_dims(image_array, axis=0)
```

```python
# show the image
print("Original Image")
plt.imshow(np.uint8(original_image))
plt.axis('off')
plt.show()
```

```python
# Create an ImageDataGenerator for augmentation
data_generator = ImageDataGenerator(rotation_range=40,
                                    width_shift_range=40,
                                    height_shift_range=0.2,
                                    shear_range=0.2,
                                    zoom_range=0.2,
                                    horizontal_flip=True,
                                    fill_mode='nearest')
```

```python
print('Augmented Images:')
plt.figure(figsize=(10, 10))

for i, batch in enumerate(data_generator.flow(image_array, batch_size=1)):
  plt.subplot(3, 3, i+1)
  augmented_image = np.uint8(batch[0])
  plt.imshow(augmented_image)
  plt.axis('off')
  if i >= 8: # display 9 images
    break;

plt.tight_layout()
plt.show()

```
