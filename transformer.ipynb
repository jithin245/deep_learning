```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
result = classifier("I hate you")[0]
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

```python
# Text Generation
```

```python
text_generator = pipeline("text-generation")
```

```python
print(text_generator("As far as I am concerned, I will", max_length=50, do_sample=True))
```

```python
print(text_generator("Once upon a time", max_length=50, do_sample=True))
```

```python
#Named Entity recogniztion
```

```python
ner_pipe = pipeline("ner")
```

```python
sequence = """ Apple is a company based in New york City. Manhattan bridge is visible from the window"""
```

```python
for entity in ner_pipe(sequence):
  print(entity)
```
